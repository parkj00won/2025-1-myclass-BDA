### KoNLPy (Korean Natural Language Processing in Python)
- **KoNLPy**는 한국어 자연어 처리를 위한 파이썬 라이브러리
  - 다양한 형태소 분석기와 텍스트 전처리 기능을 제공
  - 한국어는 형태가 복잡하고 어미 변화가 많기 때문에 영어와는 다른 접근이 필요
  - **KoNLPy**는 이러한 특징을 잘 지원하는 라이브러리 중 하나

````python
from konlpy.tag import Hannanum
han = Hannanum()
han.nouns(u'대한민국의 영토는 한반도와 그 부속도서로 한다')
````

````python
nouns = han.nouns(moon)
type(nouns)
````

````python
len(nouns)
````

````python
nouns[:10]
````

````python
nouns[-10:]
````

````python
import pandas as pd
df_word = pd.DataFrame({'word': nouns})
df_word
````

````python
df_word.info()
````

````python
df_word['word'].value_counts()
````

````python
df_word[df_word.word == '대통령']
````

````python
df_word['len'] = df_word['word'].str.len()
df_word.head()
````

````python
df_word[df_word.len >= 2]
````

````python
df_word.query('len >= 2')
````

````python
df = df_word.query('len >= 2')
df.sort_values(by='len')
````

````python
df[df.word == '대통령']
````

````python
df_freq = df.groupby('word', as_index=False) \
            .agg(n = ('word', 'count')) \
            .sort_values('n', ascending=False)
df_freq
````

````python
df_freq[df_freq.word == '대통령']
````

````python
top20 = df_freq.head(20)
top20
````
